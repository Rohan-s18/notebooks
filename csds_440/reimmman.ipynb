```python
import numpy as np
import matplotlib.pyplot as plt

def charging_voltage(t):
    return 12 * (1 - np.exp(-t / (5 * 200e-3)))

def discharging_voltage(t, v_at_switch):
    return v_at_switch * np.exp(-(t - 3) / (10 * 200e-3))

time_values = np.arange(0, 9, 1)

charging_values = [charging_voltage(t) for t in time_values[:4]]
v_at_switch = charging_values[-1]
discharging_values = [discharging_voltage(t, v_at_switch) for t in time_values[3:]]

voltage_values = np.concatenate([charging_values, discharging_values])

print("| t (s) | Vc(t) (V) |")
print("|-------|-----------|")
for t, v in zip(time_values, voltage_values):
    print(f"| {t}     | {v:.4f}    |")

plt.plot(time_values, voltage_values[:9], marker='o')
plt.title('Capacitor Voltage vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Capacitor Voltage (V)')
plt.grid(True)
plt.show()
```
