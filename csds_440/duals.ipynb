## Tetrahedron

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a tetrahedron graph
tetrahedron = nx.Graph()
tetrahedron.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (3, 4)])

# Draw the tetrahedron graph
pos = {1: (0, 0), 2: (1, 0), 3: (0.5, 0.866), 4: (0.5, 0.289)}  # Position nodes for better visualization
nx.draw(tetrahedron, pos, with_labels=True, node_size=500, node_color='skyblue', font_weight='bold')
plt.title('Tetrahedron Graph')
plt.show()

# Create the dual of the tetrahedron manually
dual_tetrahedron = nx.Graph()
dual_tetrahedron.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (3, 4), (2, 4)])

# Draw the dual of the tetrahedron graph
nx.draw(dual_tetrahedron, with_labels=True, node_size=500, node_color='lightgreen', font_weight='bold')
plt.title('Dual of Tetrahedron Graph')
plt.show()
```

## Octahedron

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an octahedron graph
octahedron = nx.Graph()
octahedron.add_edges_from([
    (1, 2), (1, 3), (1, 4), (1, 5),
    (2, 3), (2, 4), (2, 6),
    (3, 5), (3, 6),
    (4, 5), (4, 6),
    (5, 6)
])

# Draw the octahedron graph
pos = {
    1: (0, 0.5), 2: (0.5, 1), 3: (1, 0.5), 4: (0.5, 0),
    5: (0, 0), 6: (0.5, 0.5)
}  # Position nodes for better visualization
nx.draw(octahedron, pos, with_labels=True, node_size=500, node_color='skyblue', font_weight='bold')
plt.title('Octahedron Graph')
plt.show()

# Create the dual of the octahedron manually
dual_octahedron = nx.Graph()
dual_octahedron.add_edges_from([
    (1, 2), (1, 3), (1, 4), (1, 5), (2, 3), (3, 4), (4, 5), (2, 5)
])

# Draw the dual of the octahedron graph
nx.draw(dual_octahedron, with_labels=True, node_size=500, node_color='lightgreen', font_weight='bold')
plt.title('Dual of Octahedron Graph')
plt.show()
```
